```python
def quick_check(df):
    print('Shape:')
    print(df.shape)
    print('-' * 30)

    print('Missing values:')
    print(df.isna().sum())
    print(df.isna().sum().sum())
    print('-' * 30)

    print('Duplicated values:')
    print(df.duplicated().sum())
    print('-' * 50)
```

```python
# Kiểm tra missing và duplicated values
quick_check(df)
```

```python
# Chuyển đổi cột 'Time' sang định dạng Datetime
df['date'] = pd.to_datetime(df['date'])
# Sắp xếp lại theo thời gian đề phòng dữ liệu bị xáo trộn (nếu chưa được sắp xếp)
df = df.sort_values(by='date')
# Set index cho cột Time
df.set_index('date', inplace=True)
```

```python
# Resampling
# Dữ liệu ETT thường ghi mỗi 15 phút (15T) hoặc 1 giờ (1H).
df = df.resample('1H').asfreq()
```

```python
# Lưu file
# Reset index để lôi 'date' từ Index về thành cột bình thường
df = df.reset_index()
df.to_csv('ETT_cleaned_simple.csv', index=False)
print("Đã lưu file: ETT_cleaned_simple.csv")
```
